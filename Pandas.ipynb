```python
def fam_label(fsize):
    if fsize == 0:
        return 'Alone'
    elif fsize>0 and fsize<=3:
        return 'Small'
    elif fsize>3 and fsize<=6:
        return 'Medium'
    else:
        return 'Large'
```

```python
fam_label(1)
```

```python
titanic['family_type'] = titanic['family_size'].apply(fam_label)
```
